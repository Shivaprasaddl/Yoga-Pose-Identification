# Object Segmentation using Pre-trained YOLO Model (1 Class)

```python
from ultralytics import YOLO

# Load a model
model = YOLO("yolo11n-seg.pt")  # load an official model
def segment(Im,Rm):
    temp = Im.copy()
    for j in range(temp.shape[2]):
        temp[:,:,j] = temp[:,:,j]*Rm[:,:,j]
    return temp
```

```python
import os
import cv2
import numpy as np

pt ="./dataset/" # Classwise data directory
sp = "./Preprocessed_Data/" # Planned Saving Directory

for cls in os.listdir(pt):
    spp = sp+cls+'/'
    os.makedirs(spp,exist_ok=True)
    c1=1
    for j in os.listdir(pt+cls):
        #results = model([pt+cls+'/'+j],verbose=False,classes=[0])  # COCO Dataset 'Person' Class you can use more class with index [0,1,2,3,....80]
        results = model([pt+cls+'/'+j],verbose=False)  
        for r in results:
            img = np.copy(r.orig_img)
            for c in r:
                b_mask = np.zeros(img.shape[:2], np.uint8)
                contour = c.masks.xy.pop().astype(np.int32).reshape(-1, 1, 2)
                _ = cv2.drawContours(b_mask, [contour], -1, (255, 255, 255), cv2.FILLED)
                mask3ch = cv2.cvtColor(b_mask, cv2.COLOR_GRAY2BGR)
                isolated = segment(r.orig_img,mask3ch//np.max(mask3ch))
                x1, y1, x2, y2 = c.boxes.xyxy.cpu().numpy().squeeze().astype(np.int32)
                cv2.imwrite(spp+cls+'_{}.jpg'.format(c1),isolated[y1:y2, x1:x2])
                c1 +=1
    print(cls," Completed..")
```
